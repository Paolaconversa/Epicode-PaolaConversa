```python
election= px.data.election()
```

```python
election.head()
```

```python
election.iloc[:,1:4]
candidati=election.loc[:,["Coderre","Bergeron","Joly"]]
candidati.plot(kind="bar")
```

```python
el=px.data.election()
el.set_index("district", inplace=True)
el.sort_values("total",inplace=True)
plt.figure(figsize=[6,8])
el.total.plot(kind="barh")
plt.yticks(fontsize=8)
plt.grid(axis="x");
plt.xticks(np.arange(min(el.total),max(el.total)+1, 500), rotation=90);
plt.box(False)
plt.savefig("elezioni.png", dpi=600);
```

```python
el.iloc[:4,:3].plot(kind="bar", stacked=True)
```
